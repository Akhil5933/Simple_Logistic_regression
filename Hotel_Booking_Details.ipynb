```python
#Label encoding is a common technique used in data preprocessing, especially in machine learning workflows, for the conversion of category values to numerical values for training model.
#below are the categorical columns used for conversion into numeric types.
ds['hotel']=ds['hotel'].astype('category')
ds['hotel']=ds['hotel'].cat.codes

ds['arrival_date_month']=ds['arrival_date_month'].astype('category')
ds['arrival_date_month']=ds['arrival_date_month'].cat.codes

ds['meal']=ds['meal'].astype('category')
ds['meal']=ds['meal'].cat.codes

ds['country']=ds['country'].astype('category')
ds['country']=ds['country'].cat.codes

ds['market_segment']=ds['market_segment'].astype('category')
ds['market_segment']=ds['market_segment'].cat.codes

ds['distribution_channel']=ds['distribution_channel'].astype('category')
ds['distribution_channel']=ds['distribution_channel'].cat.codes

ds['deposit_type']=ds['deposit_type'].astype('category')
ds['deposit_type']=ds['deposit_type'].cat.codes

ds['customer_type']=ds['customer_type'].astype('category')
ds['customer_type']=ds['customer_type'].cat.codes

ds['reservation_status']=ds['reservation_status'].astype('category')
ds['reservation_status']=ds['reservation_status'].cat.codes


#we have to string of object type and we have to convert it to the date_time format
ds['reservation_status_date'] = pd.to_datetime(ds['reservation_status_date'])
```

```python
#As the column type present in datetime, it means that is not present integer format.
#we have to convert it into integer.
# Custom function to convert datetime to custom integer representation
def custom_int_representation(dt):
    return dt.year + dt.month  # Combine year and month into a single integer

# Apply custom function to 'datetime' column
ds['reservation_status_date'] = ds['reservation_status_date'].apply(custom_int_representation)

# showing the DataFrame
print(ds)
```

# Data Splitting

```python
#splitting the columns into dependent and independent variables
#x--independent variable
#y--dependent variable

x=ds.drop(['is_canceled'],axis=1)
y=ds[['is_canceled']]
```

```python
#It is a fundamental tool in machine learning used for splitting a dataset into training and testing sets. This process is essential for model training, validation, and evaluation.
from sklearn.model_selection import train_test_split
```

```python
#The primary purpose of train_test_split is to create two separate datasets:
#------Training Dataset: Used to train the machine learning model.
#------Testing Dataset: Used to evaluate the trained model's performance on unseen data.
#x--The feature matrix containing input features used to train and test the model.
#y--The target variable (or label) that the model aims to predict.
#test_size-----Common values range from 0.2 (20% for testing). The remaining data is used for training.
# random_state-----optional parameter used for getting testing data of equal size for every time.
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#x_train typically represents the training set of features (independent variables) in a machine learning mode
x_train
```

```python
#y_train typically represents the training set of target variables (dependent variables or response variables) in a machine learning model
y_train
```

```python
#These features represent the input data that the model will use to make predictions
x_test
```

```python
#y_test typically represents the testing set of target variables (dependent variables or response variables) in a machine learning model
y_test
```

# Model Training

```python
#This line imports the Linear Regression model from scikit-learn's linear_model module. Linear Regression is a supervised learning algorithm used for regression tasks, where the goal is to predict a continuous numerical value
from sklearn.linear_model import LinearRegression
#This line initializes a Linear Regression model object named model
model=LinearRegression()
#This line trains the Linear Regression model using the training data.
model.fit(x_train,y_train)
```

```python
x_test
```

# Predicting Score

```python
#This method takes the testing set of features (x_test) as input and returns the predicted output values based on the learned patterns from the training data.
y_pred=model.predict(x_test)
y_pred
```

```python
#indicates that the model perfectly predicts the target variable.
```

# Model Evaluation

```python
#The model.score(x_test, y_test) method in scikit-learn calculates the accuracy or R2(coefficient of determination) score of a machine learning model when evaluated on the testing data (x_test and y_test)
model.score(x_test,y_test)
```

```python
#If model is a regression model (such as Linear Regression), model.score(x_test, y_test) returns the R2score, also known as the coefficient of determination.
#R2 = 1 indicates that the model perfectly predicts the target variable.
#R2= 0 indicates that the model does not explain any variability in the target variable beyond the mean.
#R2 < 0 suggests that the model is performing worse than predicting the mean value of the target variable.
from sklearn.metrics import mean_squared_error, r2_score
mse=mean_squared_error(y_test,y_pred)
r_squared = r2_score(y_test, y_pred)
```

# Interpretation of Results

```python
#for printing the values to know the model how best it is performing
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared: {r_squared:.2f}")
```

```python
#from above score we can say the model is predicting well and good 
```

# Training Another model

```python
#choosing another model
#It is a imbalanced dataset and target variable is of binary format
#we an also consider the other model(LogisticRegression)
```

```python
##This line imports the Logistic Regression model from scikit-learn's linear_model module. Logistic Regression is a supervised learning algorithm used for regression tasks, where the goal is to predict a continuous numerical value
from sklearn.linear_model import LogisticRegression
```

```python
#This line initializes a Logistic Regression model object named lr

lr=LogisticRegression()

#This line trains the Logistic Regression model using the training data.

lr.fit(x_train,y_train)
```

# Predicting Model

```python
##This method takes the testing set of features (x_test) as input and returns the predicted output values based on the learned patterns from the training data.

y_pred_lr = lr.predict(x_test)
```

# Model Score

```python
#This line imports the accuracy_score metric from scikit-learn's metrics module. 
#accuracy_score is used to compute the accuracy of a classification model.
from sklearn.metrics import accuracy_score
#This line uses the trained Logistic Regression model (lr) to predict the target values (y_pred_test) based on the features in the testing data (x_test).
y_pred_test = lr.predict(x_test)


#This line calculates the accuracy of the model on the testing data. It compares the predicted values (y_pred_test) with the actual target values (y_test) and computes the percentage of correct predictions.
lr_accuracy_test = accuracy_score(y_test, y_pred_test)
#The accuracy score is then rounded to two decimal places and printed as a percentage.
print(f"Test Accuracy: {round(lr_accuracy_test * 100,2)}%")

#This line uses the trained Logistic Regression model (lr) to predict the target values (y_pred_train) based on the features in the training data (x_train).
y_pred_train = lr.predict(x_train)

#This line calculates the accuracy of the model on the training data. It compares the predicted values (y_pred_train) with the actual target values (y_train) and computes the percentage of correct predictions.
lr_accuracy_train = accuracy_score(y_train, y_pred_train)
#The accuracy score is then rounded to two decimal places and printed as a percentage.
print(f"Train Accuracy: {round(lr_accuracy_train * 100,2)}%")
```

# Interpretation Of Results

```python
#The training accuracy represents how well the model fits the training data.
#a very high training accuracy (almost close to 100%) can also be a sign of overfitting, where the model memorizes the training data but may not generalize well to new, unseen data.
```

```python
#The testing accuracy represents how well the model generalizes to new, unseen data (testing data).
#Ideally, you want the testing accuracy to be high, indicating that the model performs well on data it hasn't seen during training.
```

```python
#On the other hand, if both the training and testing accuracies are similar and reasonably high, it suggests that the model is performing well and can generalize to unseen data effectively.
```
